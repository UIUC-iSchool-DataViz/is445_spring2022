```python
import PIL.Image as Image
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
im = Image.open(fn).convert('RGB')
im
```

```python
im_data = np.array(im)
```

```python
im_data.shape
```

```python
fig, ax = plt.subplots(1,4,figsize=(15,3))

ax[0].imshow(im_data)
ax[1].imshow(im_data[:,:,0], cmap='Reds')
ax[2].imshow(im_data[:,:,1], cmap='Greens')
ax[3].imshow(im_data[:,:,2], cmap='Blues')

plt.show()
```
