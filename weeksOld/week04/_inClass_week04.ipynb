Use a plotting style "context" for a plot with a "with" statement:

```python
#plt.style.use('ggplot')

with plt.style.context('ggplot'):
    fig, ax = plt.subplots(1,1, figsize=(4,2))

    ax.plot(stats["max"], marker='.', linewidth=1.0, label="Max")
    ax.plot(stats["min"], marker='.', linewidth=1.0, label="Min")
    ax.set_label('Year')
    ax.set_ylabel("Square Footage")
    ax.set_yscale("log")

    ax.legend()

plt.show()
```

Use this "with" statement to build a function around the plot style:

```python
def make_plot(style): # function!  its needs a colon!
    with plt.style.context(style):
        fig, ax = plt.subplots(1,1, figsize=(4,2))
        ax.plot(stats["max"], marker='.', linewidth=1.0, label="Max")
        ax.plot(stats["min"], marker='.', linewidth=1.0, label="Min")
        ax.set_label('Year')
        ax.set_ylabel("Square Footage")
        ax.set_yscale("log")
        ax.legend()
    plt.show()
```

Now we are going to use ipywidgets to do some interactive things with our plotting function:

```python
import ipywidgets
```

```python
@ipywidgets.interact(style = plt.style.available)
def make_plot(style): # function!  its needs a colon!
    with plt.style.context(style):
        fig, ax = plt.subplots(1,1, figsize=(4,2))
        ax.plot(stats["max"], marker='.', linewidth=1.0, label="Max")
        ax.plot(stats["min"], marker='.', linewidth=1.0, label="Min")
        ax.set_label('Year')
        ax.set_ylabel("Square Footage")
        ax.set_yscale("log")
        ax.legend()
    plt.show()
```

Going to use widgets to investigate matplotlib colormaps!

```python
import matplotlib.cm as cm
```

```python
# list these colormaps:
plt.colormaps()
```

```python
# plotting matplotlib colormaps as a function:
def plot_colormaps(cmap): # input is colormap name
    fig, ax = plt.subplots(figsize=(4, 0.4))
    # this stuff below is fine if you get like 30% of it!
    col_map = plt.get_cmap(cmap) # actual colors
    matplotlib.colorbar.ColorbarBase(ax, cmap=col_map, orientation = 'horizontal')
    plt.show()
```

```python
# plot ALL COLORMAPS!!!
for cmap_id in plt.colormaps(): # looping over all colormaps
    print(cmap_id) # printing out id
    plot_colormaps(cmap_id) # showing the colormap
```

```python
@ipywidgets.interact(cmap = plt.colormaps())
def plot_colormaps(cmap): # input is colormap name
    fig, ax = plt.subplots(ncols=1,nrows=1,figsize=(4, 0.4))
    # this stuff below is fine if you get like 30% of it!
    col_map = plt.get_cmap(cmap) # actual colors
    matplotlib.colorbar.ColorbarBase(ax, cmap=col_map, orientation = 'horizontal')
    plt.show()
```

```python
import matplotlib as mpl
color_steps = np.linspace(0,1.0,256) # never changes, can put outside

@ipywidgets.interact(colormap_name = plt.colormaps())
def plot_colormap(colormap_name):
    fig = plt.figure() # create a figure object, but don't specify axes
    
    col_map = plt.get_cmap(colormap_name) # for both plots
    colors = col_map(color_steps)[:,:3] # need to re-derive both times

    # order here is: left, bottom, width, height
    ax = fig.add_axes([0.0, 0.15, 1.0, 0.8]) # first set of axis: will hold color-levels plot

    # this axes object will hold our colorlines:
    ax.plot(color_steps, colors[:,0], '-r')
    ax.plot(color_steps, colors[:,1], '-g')
    ax.plot(color_steps, colors[:,2], '-b')
    ###ax.set_ylabel('Relative Strenghts of R,G,B Representation')

    ax.set_xlim(0.0, 1.0) # first I'm just gonna set this a little wider so the tops don't get cut off
    ax.set_ylim(0.0, 1.01)
    ax.xaxis.set_ticklabels([])
    ax.yaxis.set_ticklabels([])

    ax.xaxis.set_visible(False)
    ax.yaxis.set_visible(False) 

    # colorbar
    ax_colorbar = fig.add_axes([0.0, 0.0, 1.0, 0.2]) # right below our color lines
    mpl.colorbar.ColorbarBase(ax_colorbar, cmap=col_map, orientation = 'horizontal')
    ax_colorbar.xaxis.set_visible(False)

    plt.show()
```
