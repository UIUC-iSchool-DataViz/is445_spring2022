## Starting to work with colormaps

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.plot(stats['max'], marker='.') # note: marker is telling our audience discrete data points
ax.plot(stats['min'], marker='.') # note: marker is telling our audience discrete data points
ax.set_xlabel('Year')
ax.set_ylabel('Square Footage')
ax.set_yscale('log') # our y-scale is a large range: 10 sqrt ft up to 10^6 sqrt ft, so we want to use a log here

plt.show()
```

```python
with plt.style.context("ggplot"): # "with" a particular plotting style ONLY for this plot
    fig, ax = plt.subplots(figsize=(10,5))

    ax.plot(stats['max'], marker='.') # note: marker is telling our audience discrete data points
    ax.plot(stats['min'], marker='.') # note: marker is telling our audience discrete data points
    ax.set_xlabel('Year')
    ax.set_ylabel('Square Footage')
    ax.set_yscale('log') # our y-scale is a large range: 10 sqrt ft up to 10^6 sqrt ft, so we want to use a log here

plt.show()
```

```python
def make_plot(style):
    #with plt.style.context("default"): # "with" a particular plotting style ONLY for this plot
    with plt.style.context(style): # "with" a particular plotting style ONLY for this plot
        fig, ax = plt.subplots(figsize=(10,4))

        ax.plot(stats['max'], marker='.') # note: marker is telling our audience discrete data points
        ax.plot(stats['min'], marker='.') # note: marker is telling our audience discrete data points
        ax.set_xlabel('Year')
        ax.set_ylabel('Square Footage')
        ax.set_yscale('log') # our y-scale is a large range: 10 sqrt ft up to 10^6 sqrt ft, so we want to use a log here

    plt.show()
```
