```python
# lets plot the maximum square footage per year
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(stats['max'], marker='.')
ax.set_xlabel('Year')
ax.set_ylabel("Square Footage")

plt.show()
```

```python
# lets plot the maximum square footage per year
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(stats['max'], marker='.')
ax.set_xlabel('Year')
ax.set_ylabel("Square Footage")
ax.set_yscale('log')

plt.show()
```

```python
# lets plot the maximum square footage per year
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(stats['min'], marker='.')
ax.set_xlabel('Year')
ax.set_ylabel("Square Footage")
ax.set_yscale('log')

plt.show()
```

```python
# lets plot the maximum square footage per year
fig, ax = plt.subplots(figsize=(8,3))

ax.plot(stats['max'], marker='.', label='Max')
ax.plot(stats['min'], marker='.', label='Min')
ax.set_xlabel('Year')
ax.set_ylabel("Square Footage")
ax.set_yscale('log')
ax.legend()
#fig.legend()

plt.show()
```

```python
with plt.style.context("ggplot"):
    fig, ax = plt.subplots(figsize=(8,3))
    ax.plot(stats['max'], marker='.', label='Max')
    ax.plot(stats['min'], marker='.', label='Min')
    ax.set_xlabel('Year')
    ax.set_ylabel("Square Footage")
    ax.set_yscale('log')
    ax.legend()
plt.show()
```

```python
# step 1: hard code something
with plt.style.context("ggplot"):
    fig, ax = plt.subplots(figsize=(8,3))
    ax.plot(stats['max'], marker='.', label='Max')
    ax.plot(stats['min'], marker='.', label='Min')
    ax.set_xlabel('Year')
    ax.set_ylabel("Square Footage")
    ax.set_yscale('log')
    ax.legend()
plt.show()
```

```python
# step 2: put into function
def make_plot(style):
    #with plt.style.context("ggplot"):
    with plt.style.context(style):
        fig, ax = plt.subplots(figsize=(8,3))
        ax.plot(stats['max'], marker='.', label='Max')
        ax.plot(stats['min'], marker='.', label='Min')
        ax.set_xlabel('Year')
        ax.set_ylabel("Square Footage")
        ax.set_yscale('log')
        ax.legend()
    plt.show()
```

```python
# step 3: decorator
@ipywidgets.interact(style=plt.style.available)
def make_plot(style):
    #with plt.style.context("ggplot"):
    with plt.style.context(style):
        fig, ax = plt.subplots(figsize=(8,3))
        ax.plot(stats['max'], marker='.', label='Max')
        ax.plot(stats['min'], marker='.', label='Min')
        ax.set_xlabel('Year')
        ax.set_ylabel("Square Footage")
        ax.set_yscale('log')
        ax.legend()
    plt.show()
```
