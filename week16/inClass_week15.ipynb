Keeping in mind we may or maynot want to take out (1) punctuation and (2) capitalization in general, keeping that in mind, we'll do a bit of this cleaning for our dataset.

We'll use the nltk package for most of this:
 * takes care of capitalization
 * also different "versions" of words
 
Side note: if you are into NLP (natural language processing) you can use the nltk package to get ngrams and such.

```python
word_tokens = nltk.word_tokenize(" ".join(myWords))
```

```python
word_tokens
```

Now with this different representation of words, we'll re-do our filtering process:

```python
less_words = [w for w in word_tokens if not w.lower() in stop_words]
# cleaning out of stop words again
```

```python
less_words
```

In addition to filtering out stop words we don't include punctuation.

```python
string.punctuation
```

```python
print(string.punctuation)
```

Making some choices:
 1. exclude any single punctuation marks
 1. exclude any trailing punctutation marks

```python
no_punc_words = []
for word in less_words:
    if word[-1] in string.punctuation: # last (or only!) element of word is a punctuation mark
        word = word[:-1] # take out all but last element
    if len(word) > 0: # make sure I am not left with an empty string
        no_punc_words.append(word)
```

```python
#no_punc_words
```

Remake our words dictionary with our tokenized and cleaned dataset:

```python
wordsDict = {}
for word in no_punc_words:
    if word in wordsDict:
        wordsDict[word] += 1
    else:
        wordsDict[word] = 1
```

```python
#wordsDict
```

Reading in some facebook data:

```python
filename = 'facebook_combined_sm000090_000010.txt' # sparse
filename = 'facebook_combined_sm000030_000000.txt' # more centralized

# network that includes the 10th most connected person all the way down to the 90th most connected person
```

```python
import pandas as pd
```

```python
network = pd.read_csv('/Users/jillnaiman/Downloads/'+filename, sep=' ', names=['ind1', 'ind2'])
```

```python
network
```

```python
network = network.drop_duplicates()
```

```python
network
```

```python
node_data = []
link_data = []
color_data = []
```

```python
import numpy as np
```

How many unique nodes do we have?

```python
u_nodes = np.unique( np.append(network['ind1'], network['ind2']))
```

```python
u_nodes
```

Generate our node_data:

```python
for un in u_nodes:
    node_data.append({'label':str(un), "shape":"circle"})
```

```python
#node_data
```

Specify how a particular node is connected to another:

```python
for iu, un in enumerate(u_nodes):
    #print(iu,un)
    target_ids = network.loc[network['ind1']==un] # subset of my data that includes this particular node
    #print(target_ids)
    #print(' ')
    tnodes = np.unique(target_ids['ind2'].values)
    #print(tnodes)
    #print(' ')
    for t in tnodes: # target node numbers
        tind = np.where(t == u_nodes)[0][0]
        link_data.append({'source':iu, 'target':tind})
    color_data.append('cyan')
```

```python
#link_data
```

```python
graph = bqplot.Graph(node_data=node_data, 
                    link_data=link_data,
                    colors=color_data)

# change some of the calculation parameters
graph.charge = -100
graph.link_type='line'
graph.directed=False

fig = bqplot.Figure(marks=[graph])

fig.layout.min_width='1000px'
fig.layout.min_height='900px'

fig
```
