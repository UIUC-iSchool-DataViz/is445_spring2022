How many bins should we use? Our color range in each channel goes from 0->255 -- total of 256 possible bins for each color channel.

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.hist(im_data[:,:,0].flatten(), bins=256) # all columns, all rows, 0th or first channel = Red channel

plt.show()
```
